```python
data = sns.load_dataset("iris")
data.head()
```

```python
data['species'].value_counts()
```

```python
# DRAW LINE PLOT
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

# sns.despine()

plt.show()
```

```python
# DRAW LINE PLOT
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

sns.despine()

plt.title("Title using Matplotlib Function")

plt.show()
```

```python
# Setting Limits
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

plt.xlim(5)

plt.show()

```

```python
# Setting Limits
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

plt.ylim(3)

plt.show()

```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

sns.set_style("white")

plt.show()
```

```python
def plot():
    sns.lineplot(x='sepal_length', 
                y='sepal_width',
                data=data)
```
