1. Use a for loop to calculate the number of customers that are getting the tech support and are male senior citizen.

```python
import pandas as pd

cus_ch = pd.read_csv (r'P:\DA DS AI\Data Science\Assignments\Module-5-Data-Manipulation-Using-Pandas\customer_churn.csv')
cus_ch
```

```python
#Use a for loop to calculate the number of customers that are getting Technical Support and are male senor citizen: 
count = 0 # for counting the numbers of customers

# using for loop:-
for i in range(len(cus_ch)):
    if ((cus_ch.TechSupport[i]== "Yes") &  (cus_ch.gender[i] == "Male") & (cus_ch.SeniorCitizen[i]== 1)):
        count = count + 1
print(count)
```
