```python
import pandas as pd

cus_ch = pd.read_csv (r'P:\DA DS AI\Data Science\Assignments\Module-5-Data-Manipulation-Using-Pandas\customer_churn.csv')
cus_ch
```

```python
Gender = ['Female']
  
# selecting rows based on condition
result1 = cus_ch[(cus_ch['tenure'] > 50) &
          cus_ch['gender'].isin(Gender)]
result1
```

```python
Gender = ['Male']
  
# selecting rows based on condition
result2 = cus_ch[(cus_ch['SeniorCitizen'] == 0) &
          cus_ch['gender'].isin(Gender)] 
result2
```
